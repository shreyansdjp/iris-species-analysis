```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("./iris/iris.csv")
statistical_analysis = df.describe()
df
```

```python
iris_setosa = df[df['Species'] == 'Iris-setosa'].drop(columns=['Species'])
iris_versicolor = df[df['Species'] == 'Iris-versicolor'].drop(columns=['Species'])
iris_virginica = df[df['Species'] == 'Iris-virginica'].drop(columns=['Species'])
```

```python
iris_setosa.describe()
```

```python
iris_setosa['SepalLengthCm'].hist(by=df['Species'], bins=20, figsize=(10, 6))
plt.title('Sepal Length Distribution by Species')
plt.xticks(rotation=0)

iris_setosa['SepalWidthCm'].hist(by=df['Species'], bins=20, figsize=(10, 6))
plt.title('Sepal Width Distribution by Species')
plt.xticks(rotation=0)

iris_setosa['PetalLengthCm'].hist(by=df['Species'], bins=20, figsize=(10, 6))
plt.title('Petal Length Distribution by Species')
plt.xticks(rotation=0)

iris_setosa['PetalWidthCm'].hist(by=df['Species'], bins=20, figsize=(10, 6))
plt.title('Petal Width Distribution by Species')
plt.xticks(rotation=0)
plt.show()
```

```python
categories = df['Species'].unique()

plt.scatter(iris_setosa["SepalWidthCm"], iris_setosa["SepalLengthCm"], label='Iris-setosa', color='blue')
plt.scatter(iris_versicolor["SepalWidthCm"], iris_versicolor["SepalLengthCm"], label='Iris-versicolor', color='green')
plt.scatter(iris_virginica["SepalWidthCm"], iris_virginica["SepalLengthCm"], label='Iris-virginica', color='red')
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Sepal Length (cm)")
plt.title("Sepal Width vs Sepal Length for Iris Setosa")
plt.legend()
plt.show()
```
