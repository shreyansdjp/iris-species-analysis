```python
import math
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv("./iris/iris.csv")
statistical_analysis = df.describe()
df
```

```python
df.replace({"Species": {"Iris-setosa": 0, "Iris-versicolor": 1, "Iris-virginica": 2}}, inplace=True)
df.rename(
    columns={
        "PetalLengthCm": "petal_length",
        "PetalWidthCm": "petal_width",
        "SepalLengthCm": "sepal_length",
        "SepalWidthCm": "sepal_width",
        "Species": "species",
        "Id": "id",
    },
    inplace=True,
)
df.drop(columns=["id"], inplace=True)
iris_setosa = df[df['species'] == 0].drop(columns=['species'])
iris_versicolor = df[df['species'] == 1].drop(columns=['species'])
iris_virginica = df[df['species'] == 2].drop(columns=['species'])
```

# Linear Regression

```python
x = iris_setosa[['petal_length']]
y = iris_setosa["petal_width"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=30)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

pd.concat([y_test, pd.Series(y_pred.round(1), index=y_test.index)], axis=1)
```

```python
# Evaluate
print("R²:           ", r2_score(y_test, y_pred))
print("RMSE:         ", mean_squared_error(y_test, y_pred) ** 0.5)
print("Coefficients: ", model.coef_)    # one weight per feature
print("Intercept:    ", model.intercept_)
```

# Multivariate Regression

```python
x = df.drop(columns=["species"])
y = df["species"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=30)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

pd.concat([y_test, pd.Series(y_pred.astype(int), index=y_test.index)], axis=1)
```

```python
# Evaluate
print("R²:           ", r2_score(y_test, y_pred))
print("RMSE:         ", mean_squared_error(y_test, y_pred) ** 0.5)
print("Coefficients: ", model.coef_)    # one weight per feature
print("Intercept:    ", model.intercept_)
```
