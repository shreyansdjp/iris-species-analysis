```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("./iris/iris.csv")
statistical_analysis = df.describe()
df
```

```python
iris_setosa = df[df['Species'] == 'Iris-setosa'].drop(columns=['Species'])
iris_versicolor = df[df['Species'] == 'Iris-versicolor'].drop(columns=['Species'])
iris_virginica = df[df['Species'] == 'Iris-virginica'].drop(columns=['Species'])
```

```python
lr_df = iris_setosa.drop(columns=["Id", "PetalLengthCm", "PetalWidthCm"])
lr_df.rename(
    columns={"SepalLengthCm": "sepal_length", "SepalWidthCm": "sepal_width", 'Id': 'id'},
    inplace=True,
)

training_df = lr_df[lr_df.index < lr_df.shape[0] * 0.8].reset_index(drop=True)
test_df = lr_df[lr_df.index >= lr_df.shape[0] * 0.8].reset_index(drop=True)
```

```python
# x = sepal_length, y = sepal_width
sepal_length_mean = training_df["sepal_length"].mean()
sepal_width_mean = training_df["sepal_width"].mean()

slope = (
    (
        (training_df["sepal_length"] - sepal_length_mean)
        * (training_df["sepal_width"] - sepal_width_mean)
    )
).sum() / (((training_df["sepal_length"] - sepal_length_mean) ** 2)).sum()

constant = sepal_width_mean - slope * sepal_length_mean
```

```python
pred_df = (slope * test_df['sepal_length'] + constant).round(1)
final_pred_df = pd.concat([pred_df, test_df["sepal_width"]], axis=1)
final_pred_df
```
